```python
#문제 1번
def mergeSort(A):
    if len(A) <= 1:
        return A
    mid = len(A) // 2
    left = mergeSort(A[:mid])
    right = mergeSort(A[mid:])

    merged_arr = []
    l = h = 0
    while l < len(left) and h < len(right):
        if left[l] < right[h]:
            merged_arr.append(left[l])
            l += 1
        else:
            merged_arr.append(right[h])
            h += 1
    merged_arr += left[l:]
    merged_arr += right[h:]
    return merged_arr

#quickSort
def quick_sort(A, p, r):
    if p < r:
        q = partition(A, p, r)
        quick_sort(A, p, q-1)
        quick_sort(A, q+1, r)

def partition(A, p, r):
    x=A[r]
    i = p-1
    for j in range(p, r):
        if A[j] <= x:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i+1], A[r] = A[r], A[i+1]
    return i+1
```

```python
#문제6
n = int(input())
numbers = [] # 정사각형 리스트
testing_array = []#작은 사각형으로 나누기 위한 리스트
answer = [] #마지막 4개로 이루어진 사각형 리스트
for i in range(n):
    numbers.append(list(map(int,input().split())))

cnt = 0#슬라이싱 범위를 정하기 위한 카운트

while len(answer) != 4:
    if cnt == 0: #슬라이싱 범위 정하기
        start1 = start2 = 0
        end1 = end2 = n//2
    elif cnt == 1:
        start1 = 0
        end1 = start1 = n//2
        end2 = n
    elif cnt == 2:
        start1 = end2 = n//2
        end1 = n
        start2 = 0
    elif cnt == 3:
        start1 = start2 = n//2
        end1 = end2 = n

    for i in range(start1,end1): # 정사각형 나눠서 정렬 후
        for j in range(start2,end2):
            testing_array.append(numbers[i][j])
    quick_sort(testing_array, 0, len(testing_array)-1)
    answer.append(testing_array[-2]) # 두번째로 큰값 정답리스트에 append
    cnt += 1
quick_sort(answer, 0, len(answer)-1) # 마지막 값이 4개 들어있는 사각형 정렬 후 print
print(answer[-2])
```
